Markdown cells allow you to have:

- bulleted lists,
- **bold** or *italic* text,
- `code snippets formatted in a mono-spacing font`,
- and designate section headings.

To achieve the different formatting styles, here is a brief guide:

- To create a bulleted list, start each line with a dash `-`, followed by a space.

    - Hit `tab` to indent one level.

- For *italic* text, surround the words with one asterisk `*`, on either side.

- For **bold** text, surround the words with two asterisks `**`, on either side.

- For `embedded code`, surround the words with backticks `` ` `` , on either side.

- For headings, preface the line with one or more pound signs `#`.

    - The heading will start at the largest size, and get progressively smaller with each addition `#`.

    - The `#`s must always be immediately followed by a space for the heading to be correctly interpreted.
    
    - There are five heading levels.

# One `#`s 
## Two `#`s
### Three `#`s
#### Four `#`s
##### Five `#`s

To use any of the special markdown characters in regular text, without affecting the formatting, escape them by using a backslash `\` to tell markdown to ignore them, like so: `\#`.

> *Note that to make the backslash in that example visible, it also had to be escaped.*
